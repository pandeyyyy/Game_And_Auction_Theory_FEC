First Price Sealed Bid Auction

```python
import numpy as np
import pandas as pd
import random
import math
```

```python
#############################
# Enter the number of players
n = 10

index = []
for i in range (1, n+1):
  index.append(i)
df = pd.DataFrame(index = index)
```

```python
random_data_1 = [random.randint(100, 500) for i in range (1, n+1)]
random_data_2 = [random.randint(100, 500) for i in range (1, n+1)]
df['r1_b1'] = random_data_1
df['r1_b2'] = random_data_2
df['r1_avg'] = (df['r1_b1'] + df['r1_b2'])/2
df.head(n)
```

```python
##############################
# Calculating number of rounds
m = math.sqrt(df['r1_avg'].max() - df['r1_avg'].min())
no_of_rounds = max(math.floor(m), 5)
print('Number of Rounds to be played in the game = ', no_of_rounds)
```

```python
for i in range (1, no_of_rounds):
  df['r' + str(i+1) + '_b1'] = [random.randint(100, 500) for i in range (1, n+1)]
  df['r' + str(i+1) + '_b2'] = [random.randint(100, 500) for i in range (1, n+1)]
  df['r' + str(i+1) + '_avg'] = [random.randint(100, 500) for i in range (1, n+1)]
df.head(n)
```

```python
df['r_avg'] = df.sum(axis = 1)/(3 * no_of_rounds)
df.head(n)
```

```python
max_bid = df['r_avg'].max()
win_ind = df.index[df['r_avg']==max_bid].tolist()
```

```python
##########################
# Condition for tie
j = no_of_rounds
if(len(win_ind) > 1):
  dt = df.loc[df['r_avg'] == max_bid]
  while(len(win_ind) == 1 & j > 0):
    max_bid = dt['r' + str(j) + '_avg'].max()
    win_ind = dt.index[dt[df['r' + str(j) + '_avg']==max_bid]].tolist()
    j = j-1

win_index = random.choice(win_ind)
```

```python
##########################
# Winner of game
print('Winner is player', win_index)

##########################
# How much winner will pay
print('Winner will pay an amount of $'+ str(math.floor(df.loc[win_index, 'r_avg'])))
```
