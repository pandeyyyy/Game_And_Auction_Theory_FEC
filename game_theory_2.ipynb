```python
import numpy as np
import pandas as pd
import math
import random
```

```python
#############################
# Enter the number of players
n = 20

#############################
# Number of rounds
no_of_rounds = 10

index = []
for i in range (1, n+1):
  index.append(i)

df = pd.DataFrame(index = index)
```

```python
for i in range (1, no_of_rounds + 1):
  df['r' + str(i) + '_bid'] = [random.randint(10, 100) for i in range (1, n+1)]
df.head(n)
```

```python
round_average = np.array(df.sum(axis = 0)/n).reshape(no_of_rounds, 1)
round_average
```

```python
dt = pd.DataFrame(index = index)

for i in range (0, no_of_rounds):
  dt['r' + str(i+1) + '_dev'] = df['r' + str(i+1) + '_bid'] - round_average[i]
dt
```

```python
dt['round_won'] = np.zeros(n)
dt['payoff'] = np.zeros(n)
winner = []
```

```python
for i in range (1, no_of_rounds + 1):
  round_dev = np.array(dt['r' + str(i) + '_dev']).reshape(n, 1)
  pos = np.array([n for n in round_dev if n > 0])
  max_bid = pos.min()
  win_ind = dt.index[dt['r' + str(i) + '_dev'] == max_bid].tolist()
  win_index = random.choice(win_ind)
  dt.loc[win_index, 'round_won'] = dt.loc[win_index,'round_won'] + 1
  dt.loc[win_index, 'payoff'] += math.floor(round_average[i-1])
  winner.append(win_index)
  print('Round ' + str(i) + ' is won by player', win_index)
```

```python
dt
```

```python
max_round_won = dt['round_won'].max()
win_ind = dt.index[dt['round_won'] == max_round_won].tolist()

if(len(win_ind) > 1):
  pay = np.array([dt.loc[ind, 'payoff'] for ind in win_ind]).max()
  win_ind = dt.index[dt['payoff'] == pay].tolist()

if(len(win_ind) > 1):
  for elem in win_ind:
   if elem in winner:
       win_ind = elem
       break
win_index = win_ind[0]
```

```python
df = pd.concat([df, dt], axis=1)
df
```

```python
##########################
# Winner of game
print('Winner is player', win_index)

##########################
# How much winner will pay
print('Winner will pay an amount of $'+ str(math.floor(dt.loc[win_index, 'payoff']/dt.loc[win_index, 'round_won'])))
```
